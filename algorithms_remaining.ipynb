```python
import random
from timeit import default_timer as timer
```

```python
def Ascending_Insertion_Sort(a):
    numComp=0
    for i in range(1, len(a)):
        for j in range(i, 0, -1):
            numComp += 1
            if a[j] < a[j-1]:
                temp = a[j]
                a[j] = a[j-1]
                a[j-1] = temp
            else:
                break
    return numComp
```

```python
def Descending_Insertion_Sort(a):
    numComp=0
    for i in range(1, len(a)):
        for j in range(i, 0, -1):
            numComp += 1
            if a[j] > a[j-1]:
                temp = a[j]
                a[j] = a[j-1]
                a[j-1] = temp
            else:
                break
    return numComp
```

```python
def Ascending_merge(a, lo, mid, hi):
    num_comp = 0
    # a[lo..mid] and a[mid+1..hi] are sorted
    
    # two pointers that point to the 1st element in each sorted part
    i = lo
    j = mid + 1

    # copy the original array into an auxiliary array
    aux = a.copy()

    # repeat for each slot in a[lo..hi]
    for k in range(lo, hi + 1):
        # all elements in the lower half has been merged
        if i > mid:
            a[k] = aux[j]
            j += 1
        # all elements in the upper half has been merged
        elif j > hi:
            a[k] = aux[i]
            i += 1
        # first element in the lower half is smaller than that in the upper half
        elif aux[i] < aux[j]:
            a[k] = aux[i]
            i += 1
            num_comp += 1
        else:
            a[k] = aux[j]
            j += 1
            num_comp += 1
    return num_comp
    
def Ascending_Merge_Sort(a, lo, hi):
    first_comp = 0
    second_comp = 0
    if hi <= lo:
        return 0
    mid = lo + int((hi-lo) / 2)
    if hi-lo > 1:
        first_comp = Ascending_Merge_Sort(a, lo, mid)
        second_comp = Ascending_Merge_Sort(a, mid+1, hi)
    total = Ascending_merge(a, lo, mid, hi) + first_comp + second_comp
    return total
```

```python
def Descending_merge(a, lo, mid, hi):
    num_comp = 0
    # a[lo..mid] and a[mid+1..hi] are sorted
    
    # two pointers that point to the 1st element in each sorted part
    i = lo
    j = mid + 1

    # copy the original array into an auxiliary array
    aux = a.copy()

    # repeat for each slot in a[lo..hi]
    for k in range(lo, hi + 1):
        # all elements in the lower half has been merged
        if i > mid:
            a[k] = aux[j]
            j += 1
        # all elements in the upper half has been merged
        elif j > hi:
            a[k] = aux[i]
            i += 1
        # first element in the lower half is smaller than that in the upper half
        elif aux[i] > aux[j]:
            a[k] = aux[i]
            i += 1
            num_comp += 1
        else:
            a[k] = aux[j]
            j += 1
            num_comp += 1
    return num_comp
    
def Descending_Merge_Sort(a, lo, hi):
    first_comp = 0
    second_comp = 0
    if hi <= lo:
        return 0
    mid = lo + int((hi-lo) / 2)
    if hi-lo > 1:
        first_comp = Descending_Merge_Sort(a, lo, mid)
        second_comp = Descending_Merge_Sort(a, mid+1, hi)
    total = Descending_merge(a, lo, mid, hi) + first_comp + second_comp
    return total
```

```python
def generateArray(n):
    subl=list()
    for i in range(n):
        subl.append(random.randrange(n))
    return subl

original_list=list()
cnt=0

#Generates a nested array of random integers using generateArray() of array sizes 1000,2000,3000....20000  
no_of_integers=list()
def generate_nested_Array():
    global cnt
    k=100
    while k<=2000:
        original_list.append(generateArray(k))
        no_of_integers.append(k)
        k+=100
        cnt+=1 
        
generate_nested_Array()    

# We have created 4 copies of the original list to perform the 2 sorting algorithms on the array sorted both in ascending and descending order
```

```python
#Best Case

number_of_comparisons_insertion_sort_best_case = list()
number_of_comparisons_merge_sort_best_case = list()
cpu_time_insertion_sort_best_case = list()
cpu_time_merge_sort_best_case = list()

for i in range(cnt):
    Ascending_Insertion_Sort(original_list[i])
    
    start = timer()
    c = Ascending_Insertion_Sort(original_list[i])
    end = timer()
    number_of_comparisons_insertion_sort_best_case.append(c)
    cpu_time_insertion_sort_best_case.append((end-start)*(10**6))
    
    start = timer()
    c = Ascending_Merge_Sort(original_list[i],0,len(original_list[i])-1)
    end = timer()
    number_of_comparisons_merge_sort_best_case.append(c)
    cpu_time_merge_sort_best_case.append((end-start)*(10**6))

print(number_of_comparisons_insertion_sort_best_case)
    

    
    
```

```python
#Worst Case



number_of_comparisons_insertion_sort_worst_case = list()
number_of_comparisons_merge_sort_worst_case = list()
cpu_time_insertion_sort_worst_case = list()
cpu_time_merge_sort_worst_case = list()

for i in range(cnt):
    Descending_Insertion_Sort(original_list[i])
    temp = original_list[i]
    start = timer()
    c = Ascending_Insertion_Sort(original_list[i])

    end = timer()
    number_of_comparisons_insertion_sort_worst_case.append(c)
    cpu_time_insertion_sort_worst_case.append((end-start)*(10**6))
    
    start = timer()
    c = Ascending_Merge_Sort(temp,0,len(temp)-1)
    end = timer()
    number_of_comparisons_merge_sort_worst_case.append(c)
    cpu_time_merge_sort_worst_case.append((end-start)*(10**6))




    

    
    
```
