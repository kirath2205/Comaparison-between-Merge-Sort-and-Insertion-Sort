```python
import random
from timeit import default_timer as timer
```

```python

#Ascending order Insertion Sort

def Ascending_Insertion_Sort(a):
    numComp=0
    for i in range(0, len(a)):
        for j in range(i, 0, -1):
            numComp += 1
            if a[j] < a[j-1]:
                temp = a[j]
                a[j] = a[j-1]
                a[j-1] = temp
            else:
                break
    return numComp

```

```python
#Descending order Insertion Sort

def Descending_Insertion_Sort(a):
    numComp=0
    for i in range(0, len(a)):
        for j in range(i, 0, -1):
            numComp += 1
            if a[j] > a[j-1]:
                temp = a[j]
                a[j] = a[j-1]
                a[j-1] = temp
            else:
                break
    return numComp
```

```python
def Ascending_merge(a, lo, mid, hi):
    num_comp = 0
    # a[lo..mid] and a[mid+1..hi] are sorted
    
    # two pointers that point to the 1st element in each sorted part
    i = lo
    j = mid + 1

    # copy the original array into an auxiliary array
    aux = a.copy()

    # repeat for each slot in a[lo..hi]
    for k in range(lo, hi + 1):
        # all elements in the lower half has been merged
        if i > mid:
            a[k] = aux[j]
            j += 1
        # all elements in the upper half has been merged
        elif j > hi:
            a[k] = aux[i]
            i += 1
        # first element in the lower half is smaller than that in the upper half
        elif aux[i] < aux[j]:
            a[k] = aux[i]
            i += 1
            num_comp += 1
        else:
            a[k] = aux[j]
            j += 1
            num_comp += 1
    return num_comp
    
def Ascending_Merge_Sort(a, lo, hi):
    first_comp = 0
    second_comp = 0
    if hi <= lo:
        return 0
    mid = lo + int((hi-lo) / 2)
    if hi-lo > 1:
        first_comp = Ascending_Merge_Sort(a, lo, mid)
        second_comp = Ascending_Merge_Sort(a, mid+1, hi)
    total = Ascending_merge(a, lo, mid, hi) + first_comp + second_comp
    return total
```

```python
def Descending_merge(a, lo, mid, hi):
    num_comp = 0
    # a[lo..mid] and a[mid+1..hi] are sorted
    
    # two pointers that point to the 1st element in each sorted part
    i = lo
    j = mid + 1

    # copy the original array into an auxiliary array
    aux = a.copy()

    # repeat for each slot in a[lo..hi]
    for k in range(lo, hi + 1):
        # all elements in the lower half has been merged
        if i > mid:
            a[k] = aux[j]
            j += 1
        # all elements in the upper half has been merged
        elif j > hi:
            a[k] = aux[i]
            i += 1
        # first element in the lower half is smaller than that in the upper half
        elif aux[i] > aux[j]:
            a[k] = aux[i]
            i += 1
            num_comp += 1
        else:
            a[k] = aux[j]
            j += 1
            num_comp += 1
    return num_comp
    
def Descending_Merge_Sort(a, lo, hi):
    first_comp = 0
    second_comp = 0
    if hi <= lo:
        return 0
    mid = lo + int((hi-lo) / 2)
    if hi-lo > 1:
        first_comp = Descending_Merge_Sort(a, lo, mid)
        second_comp = Descending_Merge_Sort(a, mid+1, hi)
    total = Descending_merge(a, lo, mid, hi) + first_comp + second_comp
    return total
```

```python

#Generates an array of random integers ranging from 0 to n-1, where n is the array size
def generateArray(n):
    subl=list()
    for i in range(n):
        subl.append(random.randrange(n))
    return subl

original_list=list()
cnt=0

#Generates a nested array of random integers using generateArray() of array sizes 1000,2000,3000....20000  

def generate_nested_Array():
    global cnt
    k=100
    while k<=2000:
        original_list.append(generateArray(k))
        k+=100
        cnt+=1 

generate_nested_Array()    
print(original_list[0])
# We have created 4 copies of the original list to perform the 2 sorting algorithms on the array sorted both in ascending and descending order

list_ascending_insertion = original_list
list_ascending_merge = original_list
list_descending_insertion = original_list
list_descending_merge = original_list
    
```

```python
number_of_comparisons_insertion_sort_ascending = list()
cpu_time_insertion_sort_ascending = list()

number_of_comparisons_merge_sort_ascending = list()
cpu_time_merge_sort_ascending = list()


#The main_program_ascending function loops through the nested array consisting of array sizes 1000, 2000, 3000...20000
#performs the 

def main_Program_ascending():
    

    
        
    for i in range(cnt):
        start=timer()
        no_of_comparisons_and_cpu_time_ascending = Ascending_Insertion_Sort(list_ascending_insertion[i])
        end=timer()
        number_of_comparisons_insertion_sort_ascending.append(no_of_comparisons_and_cpu_time_ascending)
        cpu_time_insertion_sort_ascending.append(end-start)
    print('Number of comparisons made while using insertion sort algorithm in array sizes 1000,2000,3000....20000 sorted in ascending order respectively:')    
    print(number_of_comparisons_insertion_sort_ascending)
    print()
    print('The CPU time recorded in seconds while using insertion sort algorithm in array sizes 1000,2000,3000....20000 sorted in ascending order respectively:')
    print(cpu_time_insertion_sort_ascending)
    print()
    
    
    for i in range(cnt):
        start_time_merge_ascending = timer()
        c = Ascending_Merge_Sort(list_ascending_merge[i],0,len(list_ascending_merge[i])-1)
        end_time_merge_ascending = timer()
        cpu_time_merge_sort_ascending.append(end_time_merge_ascending-start_time_merge_ascending)
        number_of_comparisons_merge_sort_ascending.append(c)
    
    print('Number of comparisons made while using merge sort algorithm in array sizes 1000,2000,3000....20000 sorted in ascending order respectively: ')
    print(number_of_comparisons_merge_sort_ascending)
    print()
    print('The CPU time recorded in seconds while using merge sort algorithm in array sizes 1000,2000,3000....20000 sorted in ascending order respectively:')
    print(cpu_time_merge_sort_ascending)
    

main_Program_ascending()    
      
        
        
```
